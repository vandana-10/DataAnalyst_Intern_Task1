#Exploratory Data Analysis (EDA)

Exploratory Data Analysis (EDA) with Python involves analyzing and summarizing data to gain insights and understand its underlying patterns, relationships, and distributions using Python programming language.


**Importing all Necessary file for Exploratory Data Analysis (EDA)**

**# Handling Duplicates**

```python
# In titanic dataset there are NO duplicate records in the dataset

data_dup = data[data.duplicated()]
data_dup
```

**Removing Duplicates**

```python
# There is no duplicates in dataset but I'm just showing how can we drop duplicates

data_drop = data.drop_duplicates()
data_drop.tail()
```

```python
# if we want to remove duplicates fom specific columns we use Subset function to drop the column duplicates
data_drop = data.drop_duplicates(subset='PassengerId')
data_drop.tail()
```

**Checking Null Values**

```python
data.isnull().sum()


# isnull() function helps to check NULL values in dataset
# sum gives the count of null values in a Column
```

```python
data.shape
```

```python
# In Cabin column has lot of NAN values so, I dropped the entire column
# Drop() function helps to remove the column permanently where axis = 1 specifics that your'e dropping a column

data = data.drop('Cabin',axis = 1)
data.head()
```

```python
data['Age']
```

```python
# distplot is used to check the Distrbution of Age Column
sns.displot(data['Age'])
```

```python
# age is a Numerical column so we can fill null values use Mean/Median/Mode (Statstical Value)
# For Categorical Values we use ffill/bfill

data['Age'] = data['Age'].fillna(data['Age'].median())
data['Age']
```

```python
# Now I Handled all the null values and Dataset is more efficient and Understandable.

data.isnull().sum()
```

```python
# Fare column has one Null Values So I use dropna() function to remove the Null Value.

data = data.dropna()
data.isnull().sum()
```

# Checking the Outliers


```python
sns.boxplot(data['Fare'])
```

```python
data['Fare'].describe()
```

** Treating Outliers using Inter Quartile Range(IQR)**

```python
# Treating Outliers using Inter Quartile Range(IQR)
#IQR = Q3-Q1
IQR = 31.50 - 7.895
IQR
```

```python
upper_fence = 31.5+(1.5*23.6)
upper_fence                       # Up = Q3+(1.5*IQR)
```

```python
Lower_fence = 7.89-(1.5*23.6)  # Lp = Q1-(1.5*IQR)
Lower_fence
```

```python
data['Fare'] = np.where( data['Fare']>upper_fence,upper_fence,data['Fare'])  # Removed the Outliers which are greater than Upper_fence Values
```

```python
sns.boxplot(data['Fare'])
```

```python
sns.boxplot(data['Age'])
```

```python
data['Age'].describe()
```

```python
IQR = 35.0 - 23.0
IQR
```

```python
upper_fence = 35.0+(1.5*12.0)
upper_fence
```

```python
lower_fence = 23.0- (1.5*12.0)
lower_fence
```

```python
data['Age'] = np.where( data['Age']>upper_fence,upper_fence,data['Age'])
```

```python
data['Age'] = np.where( data['Age']<lower_fence,lower_fence,data['Age'])
```

```python
sns.boxplot(data['Age'])
# Removed all the Outliers using IQR
```

```python
#The bar chart shows that females paid significantly higher ticket prices compared to males.

plt.bar(data['Sex'],data['Fare'] , color = 'Purple')
plt.xlabel(' Gender')
plt.ylabel('Ticket Price')
plt.title('Price by Gender')
plt.show()
```

```python
data.head()
```

```python
#The chart indicates that most males did not survive, while the majority of females did survive.
#This suggests a higher survival rate for females compared to males, possibly reflecting prioritization of women during rescue efforts.


sns.histplot(data = data , x ='Sex' ,bins = 2 , hue = 'Survived')
plt.title('Survival Count by Gender')
plt.show()
```

```python
sns.countplot(data=data, x='Sex', hue='Survived')
```

```python
#The chart shows that most male passengers were in the third class, while female passengers were more evenly distributed across
#the three classes. Notably, fewer females were in second class compared to first and third, suggesting different class distribution
#patterns between genders.


sns.countplot(x = "Sex", hue= 'Pclass',data= data )
```

```python
#The bar chart shows that the average age of survivors and non-survivors is nearly the same for both males and females, around 29–30 years.
#This suggests that age was not a major differentiating factor in survival outcomes across genders.

sns.barplot(data =data , y ='Age', x ='Sex', hue = 'Survived' )
plt.title("Survival By Age")
plt.show()
```

```python
#The scatter plot shows that most passengers paid lower ticket fares (under 100), regardless of age.
#However, a few individuals—mostly aged between 30 and 60—paid significantly higher fares, exceeding 200 and
#even reaching above 500, indicating the presence of premium ticket holders in that age range.


sns.scatterplot(data=data, x='Age', y='Fare')
plt.title('Ticket Fare on Age')
plt.show()
```

```python
#The line plot shows that passengers who survived (orange line) generally paid higher fares across most age groups
# compared to those who did not survive (blue line). The most significant fare differences are observed in the 30–60 age range,
#where survivors paid notably higher fares, potentially indicating a link between survival and access to higher-class accommodations.

sns.lineplot(data =data ,x ='Age',y ='Fare' , hue='Survived')
plt.title('Fare by Age (on Survived)')
plt.show()
```

```python
#Passengers who embarked from port Q had the highest survival rate, followed by those from port C,
#while those from port S had the lowest. This suggests that embarkation point may be associated with survival likelihood,
# possibly due to differences in class or ship sections accessed from each port.

sns.barplot(data = data, x= 'Embarked', y= 'Survived'  )
```

```python
sns.countplot(data = data, x= 'Embarked', hue= 'Survived'  )
plt.title('Survival on Emarked')
plt.show()
```

```python
# Histogram shows a count or Proportion of bins
# KDE(Kernel Density Estimator): Draws a smooth curve to show the distribution.

sns.histplot(data=data, x='Age', bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```
