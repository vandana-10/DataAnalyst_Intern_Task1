# **CHECKING FOR DUPLICATES**

```python
data_dup_rec=data[data.duplicated()].sum()
data_dup_rec
```

# **CHECKING NULL VALUES**

```python
data.isnull().sum()    # By this data there 6 columns which contain NULL values so we can Replace thier values .
```

```python
data['director'].replace(np.nan, 'No Director',inplace=True)
data['cast'].replace(np.nan , "No Cast",inplace=True)
```

**I use Mode(Categorical Values) for Country and Rating Columns  to Replace the Null Values in Data Set**

```python
data.country.mode()
```

```python
data['country'].replace(np.nan,"United States" ,inplace=True)
```

```python
data.rating.mode()
```

```python
data['rating'].replace(np.nan,"TV-MA" ,inplace=True)
```

```python
data['date_added'].replace(np.nan,"No Date",inplace=True)
```

**I dropped the Remaining columns which has less null Values Or We can Replace the Values Like Other Columns but NULL values are very Low So I dropped them.**

```python
data = data.dropna()
data.isnull().sum()
```
